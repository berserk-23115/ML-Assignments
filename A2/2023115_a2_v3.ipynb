```python
class BaseDataset(ABC):
    """
    Shared interface for modality-specific datasets.

    Attributes
    - modality: string describing modality ("image" | "text"). Set by subclasses.
    """

    def __init__(self, data_dir: str):
        self.data_dir: str = data_dir
        self.modality: str = ""

    @abstractmethod
    def __len__(self) -> int:
        ...

    @abstractmethod
    def __getitem__(self, idx: int) -> Dict[str, Any]:
        ...


class ImageDataset(BaseDataset):
    """
    Load images organized under per-class folders.
    """

    def __init__(self, path_to_image_data_dir: str = PATH_TO_IMAGE_TRAIN_DIR):
        super().__init__(path_to_image_data_dir)
        self.modality = "image"
        self.records = []

        # YOUR CODE HERE
        if os.path.exists(path_to_image_data_dir):
            classes = sorted(os.listdir(path_to_image_data_dir))
            # Filter out .DS_Store or other non-folders if necessary, though sorted listdir usually fine
            classes = [c for c in classes if not c.startswith('.')]
            
            for class_idx, class_name in enumerate(classes):
                class_path = os.path.join(path_to_image_data_dir, class_name)
                if os.path.isdir(class_path):
                    for img_name in os.listdir(class_path):
                        if img_name.startswith('.'): continue
                        img_path = os.path.join(class_path, img_name)
                        self.records.append({
                            'path': img_path,
                            'class': class_idx # Using integer encoding for class
                        })
        

    def __len__(self) -> int:
        return len(self.records)

    def __getitem__(self, idx: int) -> Dict[str, Any]:
        record = {
            "class": None,
        }
        # YOUR CODE HERE
        rec = self.records[idx]
        # CHANGE: Load as RGB to keep color features. 
        # The Sampler handles conversion to 'L' internally, so this is safe.
        img = Image.open(rec['path']).convert('RGB')
        
        record = {
            "image": img,           
            "data": np.array(img), 
            "class": rec['class']
        }
        return record
        

class TextDataset(BaseDataset):
    """
    Load text data.
    """

    def __init__(self, path_to_text_data_dir: str = PATH_TO_TEXT_TRAIN_DIR):
        super().__init__(path_to_text_data_dir)
        self.modality = "text"
        self.records = []

        # YOUR CODE HERE
        files = glob.glob(os.path.join(path_to_text_data_dir, '*.jsonl'))
        for file_path in files:
            with open(file_path, 'r') as f:
                for line in f:
                    if line.strip():
                        self.records.append(json.loads(line))
        

    def __len__(self) -> int:
        return len(self.records)

    def __getitem__(self, idx: int) -> Dict[str, Any]:
        record = {
            "class": None,
        }
        # YOUR CODE HERE
        rec = self.records[idx]
        # Combine relevant text fields
        text_parts = [
            rec.get('norm', ''),
            rec.get('situation', ''),
            rec.get('intention', ''),
            rec.get('moral_action', ''),
            rec.get('moral_consequence', '')
        ]
        combined_text = " ".join([str(p) for p in text_parts if p])
        
        record = {
            "data": combined_text,
            "class": str(rec.get('label', ''))
        }
        return record

       


    def get_corpus(self) -> List[str]:
        """
        [OPTIONAL]
        Return the full collection of text strings in the dataset.

        Requirement:
        - This function is optional and you may leave the code untouched if you do not want to use it.
        - Extract and return the raw text for each stored record.
        - Suggested use case: you can use this function to build tf-idf vectorizer later in the `get_text_features` function.
        """
        corpus = []
        # YOUR CODE HERE
        for i in range(len(self.records)):
            corpus.append(self.__getitem__(i)['data'])
        return corpus

    
```

```python
class BaseSampler(ABC):
    """Common interface for samplers"""

    @abstractmethod
    def sample_indices(self, dataset: BaseDataset, num_samples: Optional[int] = None) -> List[int]:
        ...
```

```python
class RandomSampler(BaseSampler):
    """Uniform random sampler.

    Requirement:
    - Probability of each record being sampled is uniform.
    - Sampling is without replacement.
    - Works for any dataset modality.
    """

    def __init__(self, dataset: BaseDataset) -> None:
        """
        Initialize the uniform sampler.

        Requirement:
        - No precomputation is required; may store references if needed.
        """
        # YOUR CODE HERE
        self.dataset = dataset
        

    def sample_indices(self, dataset: BaseDataset, num_samples: Optional[int] = None) -> List[int]:
        """
        Draw indices uniformly at random without replacement.

        Requirement:
        - Every record has equal probability.
        - Respect `num_samples`; if None, select all records.
        - Return a list of unique indices.
        """
        # YOUR CODE HERE
        if num_samples is None:
            num_samples = len(dataset)
        
        # Ensure we don't sample more than available
        n = min(num_samples, len(dataset))
        return random.sample(range(len(dataset)), n)
        
```

```python
class WeightedRandomSampler(BaseSampler):
    """Weighted random sampler favoring least-similar classes.

    Requirement:
    - Derive class similarity internally (no external information):
      - Image: cosine similarity of grayscale histograms.
      - Text: cosine similarity of tokenized sentence vectors.
    - Convert similarity to weights that prefer less-similar classes (inverse similarity).
    - This mapping from class to weight should be stored in the `class_weights` attribute.
    - Sampling is 
    """

    def __init__(self, dataset: BaseDataset) -> None:
        """
        Initialize and compute class weights from dataset-internal representations.

        Requirement:
        - Build a per-class representation per modality and compute pairwise cosine similarities.
        - Convert similarities into class-selection weights that emphasize dissimilar classes.
        - Do not use any external/precomputed similarity information.
        """
        self.class_weights = {}
        # YOUR CODE HERE

        class_indices = {}
        for idx in range(len(dataset)):
            # We access the internal records if possible to avoid loading images here
            # But the interface is dataset[idx]. We rely on __getitem__
            item = dataset[idx]
            cls = item['class']
            
            if cls not in class_indices:
                class_indices[cls] = []
            class_indices[cls].append(idx)
            
        # 2. Compute mean feature vector for each class
        class_mean_features = {}
        
        # Setup for text if needed
        corpus = []
        if dataset.modality == 'text':
            corpus = dataset.get_corpus()
            # Initialize vectorizer once
            if not hasattr(get_text_features, "vectorizer"):
                 get_text_features({'data': ''}, corpus_texts=corpus)

        for cls in class_indices:
            indices = class_indices[cls]
            features_list = []
            
            # Optimization: Use a subset (first 100) to estimate class mean 
            # to avoid loading all images during init
            subset_indices = indices[:100]
            
            for idx in subset_indices:
                item = dataset[idx]
                if dataset.modality == 'image':
                    # Grayscale Histogram
                    img = item['image'].convert('L')
                    hist = np.array(img.histogram())
                    # Normalize
                    s = np.sum(hist)
                    if s > 0: hist = hist / s
                    features_list.append(hist)
                elif dataset.modality == 'text':
                    feat = get_text_features(item, corpus_texts=corpus)
                    features_list.append(feat)
            
            if features_list:
                class_mean_features[cls] = np.mean(features_list, axis=0)

        # 3. Compute pairwise cosine similarity
        classes = list(class_mean_features.keys())
        n_classes = len(classes)
        similarity_scores = {} # dict instead of defaultdict
        
        for i in range(n_classes):
            cls_i = classes[i]
            vec_i = class_mean_features[cls_i]
            norm_i = np.linalg.norm(vec_i)
            
            sim_sum = 0.0
            for j in range(n_classes):
                if i == j: continue
                cls_j = classes[j]
                vec_j = class_mean_features[cls_j]
                norm_j = np.linalg.norm(vec_j)
                
                denom = norm_i * norm_j
                if denom == 0:
                    sim = 0.0
                else:
                    sim = np.dot(vec_i, vec_j) / denom
                sim_sum += sim
            
            similarity_scores[cls_i] = sim_sum

        # 4. Inverse similarity weights
        for cls in classes:
            # Avoid division by zero for single class case
            avg_sim = 1.0
            if n_classes > 1:
                avg_sim = similarity_scores[cls] / (n_classes - 1)
                
            self.class_weights[cls] = 1.0 / (avg_sim + 1e-9)
        

    def sample_indices(self, dataset: BaseDataset, num_samples: Optional[int] = None) -> List[int]:
        """
        Sample record indices by first selecting classes via inverse-similarity weights.

        Requirement:
        - Respect `num_samples`; if None, select all available records.
        - Choose classes according to computed weights, then sample records without replacement.
        - Return a list of unique indices, length ≤ `num_samples`.
        """
        num_samples = min(num_samples, len(dataset))
        # YOUR CODE HERE
        if num_samples is None:
            num_samples = len(dataset)
        
        # Prepare weights for np.random.choice
        classes = list(self.class_weights.keys())
        if not classes:
            return random.sample(range(len(dataset)), min(num_samples, len(dataset)))
            
        weights = [self.class_weights[c] for c in classes]
        total_w = sum(weights)
        if total_w == 0:
            probs = [1.0/len(weights)] * len(weights)
        else:
            probs = [w / total_w for w in weights]
            
        # Sample classes with replacement based on weights
        chosen_classes = np.random.choice(classes, size=num_samples, p=probs)
        
        # Group available indices by class
        available_indices = {}
        for idx in range(len(dataset)):
            # We assume class is consistent with init
            # To be fast, we rely on the fact that we can construct this mapping
            # using the dataset interface again.
            item = dataset[idx]
            cls = item['class']
            if cls not in available_indices:
                available_indices[cls] = []
            available_indices[cls].append(idx)
            
        # Shuffle indices within each class to pick randomly
        for cls in available_indices:
            random.shuffle(available_indices[cls])
            
        selected_indices = []
        for cls in chosen_classes:
            if cls in available_indices and len(available_indices[cls]) > 0:
                # pop ensures without replacement per class
                selected_indices.append(available_indices[cls].pop())
            else:
                # Fallback if a class runs out of samples? 
                # We skip to maintain "without replacement" integrity or pick from others.
                # Here we skip, and fill remaining later.
                pass
                
        # Fill if we are short (due to exhausted classes)
        if len(selected_indices) < num_samples:
            remaining = []
            for cls in available_indices:
                remaining.extend(available_indices[cls])
            
            needed = num_samples - len(selected_indices)
            if needed > 0 and remaining:
                # Take what is left
                k = min(needed, len(remaining))
                selected_indices.extend(random.sample(remaining, k))
                
        return selected_indices
```

```python
def get_image_features(sample: Dict[str, Any]) -> np.ndarray:
    """
    Get image features from a sample.

    Requirement:
    - Perform any type of feature engineering on the image as long as you do not import libraries that are not already imported.
    - You must populate the `features` variable with the features of the image.
    """
    features = None
    ### YOUR CODE HERE
    img = sample["image"]
    if img.mode != 'RGB':
        img = img.convert('RGB')

    resized = img.resize((64, 64))
    arr = np.array(resized, dtype=np.float32) / 255.0
    gray = np.dot(arr[..., :3], [0.299, 0.587, 0.114])

    feat = []

    # ══════════════════════════════════════════════
    # A: Global color histograms — 64 bins × 3 = 192
    # ══════════════════════════════════════════════
    for c in range(3):
        hist, _ = np.histogram(arr[:, :, c], bins=64, range=(0, 1), density=True)
        feat.extend(hist)

    # ══════════════════════════════════════════════
    # B: 3×3 spatial grid — color stats per zone
    #    9 zones × 3 channels × 3 stats = 81
    # ══════════════════════════════════════════════
    step = 64 // 3
    for i in range(3):
        for j in range(3):
            r0, r1 = i * step, (i + 1) * step if i < 2 else 64
            c0, c1 = j * step, (j + 1) * step if j < 2 else 64
            cell = arr[r0:r1, c0:c1, :]
            for ch in range(3):
                vals = cell[:, :, ch].ravel()
                mu = np.mean(vals)
                sd = np.std(vals) + 1e-8
                skew = np.mean(((vals - mu) / sd) ** 3)
                feat.extend([mu, sd, skew])

    # ══════════════════════════════════════════════
    # C: HOG-like gradient orientation histograms
    #    4×4 spatial grid × 9 orientation bins = 144
    #    *** This is the single most important addition ***
    # ══════════════════════════════════════════════
    padded = np.pad(gray, ((1, 1), (1, 1)), mode='edge')
    dx = padded[1:-1, 2:] - padded[1:-1, :-2]   # horizontal gradient
    dy = padded[2:, 1:-1] - padded[:-2, 1:-1]   # vertical gradient
    magnitude = np.sqrt(dx ** 2 + dy ** 2)
    angle = np.arctan2(np.abs(dy), np.abs(dx)) * 180 / np.pi  # 0–90°, unsigned

    n_bins = 9
    cell_size = 16   # 64 / 4 = 16px cells
    bin_size = 90.0 / n_bins

    for i in range(4):
        for j in range(4):
            r0, r1 = i * cell_size, (i + 1) * cell_size
            c0, c1 = j * cell_size, (j + 1) * cell_size
            cell_mag = magnitude[r0:r1, c0:c1].ravel()
            cell_ang = angle[r0:r1, c0:c1].ravel()
            hist = np.zeros(n_bins)
            for k in range(n_bins):
                mask = (cell_ang >= k * bin_size) & (cell_ang < (k + 1) * bin_size)
                hist[k] = np.sum(cell_mag[mask])
            norm = np.linalg.norm(hist) + 1e-8
            feat.extend(hist / norm)

    # ══════════════════════════════════════════════
    # D: Color saturation histogram — 32 bins = 32
    #    Helps distinguish vivid (animals) vs muted (vehicles) images
    # ══════════════════════════════════════════════
    r_ch, g_ch, b_ch = arr[:, :, 0], arr[:, :, 1], arr[:, :, 2]
    max_c = np.maximum(np.maximum(r_ch, g_ch), b_ch) + 1e-8
    min_c = np.minimum(np.minimum(r_ch, g_ch), b_ch)
    sat = (max_c - min_c) / max_c
    sat_hist, _ = np.histogram(sat, bins=32, range=(0, 1), density=True)
    feat.extend(sat_hist)

    # ══════════════════════════════════════════════
    # E: Laplacian texture histogram — 32 bins = 32
    #    Captures spot/stripe textures (giraffe, tiger, etc.)
    # ══════════════════════════════════════════════
    padded2 = np.pad(gray, ((1, 1), (1, 1)), mode='edge')
    lap = (padded2[:-2, 1:-1] + padded2[2:, 1:-1] +
           padded2[1:-1, :-2] + padded2[1:-1, 2:] - 4 * gray)
    lap_hist, _ = np.histogram(lap, bins=32, range=(-1, 1), density=True)
    feat.extend(lap_hist)

    # ══════════════════════════════════════════════
    # F: Downsampled 32×32 grayscale — 1024 features
    #    Captures rough shape/silhouette (much more than 16×16)
    # ══════════════════════════════════════════════
    small_gray = np.array(
        img.convert('L').resize((32, 32)), dtype=np.float32
    ) / 255.0
    feat.extend(small_gray.ravel())

    features = np.array(feat, dtype=np.float32)
    return features

    

def get_text_features(sample: Dict[str, Any], **kwargs) -> np.ndarray:
    """
    Get text features from a sample.
    
    Requirement:
    - Perform any type of feature engineering on the text as long as you do not import libraries that are not already imported.
    - You must populate the `features` variable with the features of the text.
    """
    features = None
    # YOUR CODE HERE

    if not hasattr(get_text_features, "vectorizer"):
        corpus = kwargs.get("corpus_texts", [])
        # Increase features to 2000-5000 for better accuracy
        vectorizer = sklearn.feature_extraction.text.TfidfVectorizer(
            max_features=2000, 
            stop_words='english',
            ngram_range=(1, 2) # Use bigrams to capture context
        )
        if corpus:
            vectorizer.fit(corpus)
        else:
            vectorizer.fit([sample['data']])
        get_text_features.vectorizer = vectorizer
    
    vec = get_text_features.vectorizer.transform([sample['data']])
    features = vec.toarray()[0]
    return features
```

```python
class DataLoader:
    """
    A modality and sampler agnostic data loader.
    """

    def __init__(self, modality: str, sampler: str, data_fraction: Optional[float] = None, is_train: bool = True) -> None:
        if is_train:
            path_to_image_dir = PATH_TO_IMAGE_TRAIN_DIR
            path_to_text_dir = PATH_TO_TEXT_TRAIN_DIR
        else:
            path_to_image_dir = PATH_TO_IMAGE_TEST_DIR
            path_to_text_dir = PATH_TO_TEXT_TEST_DIR

        self.modality = modality
        self.sampler_name = sampler
        self.data_fraction = data_fraction

        modality2dataset = {
            "image": ImageDataset(path_to_image_dir),
            "text": TextDataset(path_to_text_dir),
        }
        sampler_name2sampler_class = {
            "random": RandomSampler,
            "weighted_random": WeightedRandomSampler,
        }
        self.dataset = modality2dataset[self.modality]
        self.sampler = sampler_name2sampler_class[self.sampler_name](self.dataset)

    def get_features_and_labels(self) -> Tuple[List[np.ndarray], List[str]]:
        # if self.data_fraction is None:
        #     self.data_fraction = len(self.dataset)
        # num_samples = int(self.data_fraction * len(self.dataset))
        if self.data_fraction is None:
            num_samples = len(self.dataset)
        else:
            num_samples = int(self.data_fraction * len(self.dataset))
        indices = self.sampler.sample_indices(self.dataset, num_samples)
        X: List[np.ndarray] = []
        y: List[Any] = []
        for idx in indices:
            sample = self.dataset[idx]
            if getattr(self.dataset, 'modality', '') == 'image':
                feat = get_image_features(sample)
                label = sample["class"]
            else:
                feat = get_text_features(sample, corpus_texts=self.dataset.get_corpus())
                label = sample.get("class")
            X.append(feat)
            y.append(label)
        return X, y
```

```python
############################
# Q3: Model Trainer
############################

class ModelTrainer:
    def __init__(self, modality: str, sampler: str, data_fraction: Optional[float] = None):
        self.modality = modality
        self.sampler = sampler
        self.data_fraction = data_fraction

        self.train_dataloader = DataLoader(modality=modality, sampler=sampler, data_fraction=data_fraction, is_train=True)
        self.test_dataloader = DataLoader(modality=modality, sampler=sampler, data_fraction=None, is_train=False)

        self.train_features, self.train_labels = self.train_dataloader.get_features_and_labels()
        self.test_features, self.test_labels = self.test_dataloader.get_features_and_labels()

        # label encoding using LabelEncoder
        self.label_encoder = sklearn.preprocessing.LabelEncoder()
        self.train_labels = self.label_encoder.fit_transform(self.train_labels)
        self.test_labels = self.label_encoder.transform(self.test_labels)

        self.train_features = np.array(self.train_features)
        self.test_features = np.array(self.test_features)
        self.train_labels = np.array(self.train_labels)
        self.test_labels = np.array(self.test_labels)

    def train_classifier(self, classifier_object):
        self.classifier = classifier_object
        self.classifier.fit(self.train_features, self.train_labels)

    def evaluate_classifier(self):
        test_predictions = self.classifier.predict(self.test_features)
        self.confusion_matrix_ = sklearn.metrics.confusion_matrix(self.test_labels, test_predictions)
        accuracy = sklearn.metrics.accuracy_score(self.test_labels, test_predictions)
        self.plot_confusion_matrix()
        print(f'Accuracy for modality: {self.modality}, sampler: {self.sampler}, data fraction: {self.data_fraction} is {accuracy}')
        return accuracy

    def plot_confusion_matrix(self):
        label_names = self.label_encoder.inverse_transform(range(len(self.label_encoder.classes_)))
        plt.rcParams["figure.figsize"] = (7, 7)
        sns.heatmap(self.confusion_matrix_, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=label_names, yticklabels=label_names)
        plt.title(f'Confusion Matrix for modality: {self.modality}, sampler: {self.sampler}, data fraction: {self.data_fraction}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.tight_layout()
        plt.savefig(f'confusion_matrix_{self.modality}_{self.sampler}_{self.data_fraction}.png')

    def save_confusion_matrix(self):
        # save as numpy matrix
        np.save(f'confusion_matrix_{self.modality}_{self.sampler}_{self.data_fraction}.npy', self.confusion_matrix_)
```

# Q4 Implement Linear Regression from Scratch (Using NumPy) (2.0 Points)

1) Normal Equation
2) Gradient Descent

## Understanding the Math Behind Linear Regression

**Linear Regression Model:**
$$\hat{y} = X \cdot \beta$$

**Cost Function (Mean Squared Error):**
$$J(\beta) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}_i - y_i)^2$$

**Normal Equation (Closed-form solution):**
$$\beta = (X^T X)^{-1} X^T y$$

**Gradient Descent Update Rule:**
$$\beta := \beta - \alpha \cdot \frac{1}{m} X^T (X\beta - y)$$

```python
class LinearRegressionScratch:
    """
    Linear Regression implementation from scratch using NumPy.
    
    Supports two methods:
    1. Normal Equation (closed-form solution)
    2. Gradient Descent (iterative optimization)
    """
    
    def __init__(self, method='normal', learning_rate=0.01, n_iterations=1000):
        """
        Initialize the Linear Regression model.
        
        Parameters:
        -----------
        method : str
            'normal' for Normal Equation, 'gradient' for Gradient Descent
        learning_rate : float
            Learning rate for gradient descent (alpha)
        n_iterations : int
            Number of iterations for gradient descent
        """
        # YOUR CODE HERE
        self.method = method
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
        self.cost_history = []
        
    
    def _add_bias_term(self, X):
        """
        Add a column of ones for the bias term.
        
        Parameters:
        -----------
        X : numpy array
            Feature matrix of shape (m, n)
        
        Returns:
        --------
        X_b : numpy array
            Feature matrix with bias column, shape (m, n+1)
        """
        # YOUR CODE HERE
        m = X.shape[0]
        bias_column = np.ones((m, 1))
        X_b = np.hstack((bias_column, X))
        return X_b
        
    
    def _compute_cost(self, X, y, theta):
        """
        Compute the Mean Squared Error cost function.
        
        Parameters:
        -----------
        X : numpy array
            Feature matrix with bias term (m, n+1)
        y : numpy array
            Target values (m,)
        theta : numpy array
            Model parameters (n+1,)
        
        Returns:
        --------
        cost : float
            Mean Squared Error
        """
        # YOUR CODE HERE
        y_pred = X.dot(theta)
        cost = np.mean((y_pred - y) ** 2)
        return cost
        
    
    def _compute_gradient(self, X, y, theta):
        """
        Compute the gradient of the cost function.
        
        Parameters:
        -----------
        X : numpy array
            Feature matrix with bias term (m, n+1)
        y : numpy array
            Target values (m,)
        theta : numpy array
            Model parameters (n+1,)
        
        Returns:
        --------
        gradient : numpy array
            Gradient vector (n+1,)
        """
        # YOUR CODE HERE
        m = X.shape[0]
        y_pred = X.dot(theta)
        gradient = (1/m) * X.T.dot(y_pred - y)
        return gradient
        
    
    def fit_normal_equation(self, X, y):
        """
        Fit the model using the Normal Equation (closed-form solution).
        
        Formula: θ = (X^T X)^(-1) X^T y
        
        Parameters:
        -----------
        X : numpy array
            Feature matrix (m, n)
        y : numpy array
            Target values (m,)
        """
        # YOUR CODE HERE
        X_b = self._add_bias_term(X)
        weights = np.linalg.pinv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
        self.weights = weights[1:]
        self.bias = weights[0]
        
    
    def fit_gradient_descent(self, X, y):
        """
        Fit the model using Gradient Descent optimization.
        
        Parameters:
        -----------
        X : numpy array
            Feature matrix (m, n)
        y : numpy array
            Target values (m,)
        """
        # YOUR CODE HERE
        X_b = self._add_bias_term(X)
        lr_alpha = self.learning_rate
        feat_cnt = X.shape[1]
        sample_cnt = X.shape[0]
        weights = np.random.randn(feat_cnt + 1)  # Initialize weights randomly
        while (self.n_iterations > 0):
            y_pred = X_b.dot(weights)
            gradient = (1/sample_cnt) * X_b.T.dot(y_pred - y)
            weights -= lr_alpha * gradient
            self.cost_history.append(self._compute_cost(X_b, y, weights))
            self.n_iterations -= 1
        
        self.weights = weights[1:]
        self.bias = weights[0]
        
    
    def fit(self, X, y):
        """
        Fit the model using the specified method.
        
        Parameters:
        -----------
        X : numpy array
            Feature matrix (m, n)
        y : numpy array
            Target values (m,)
        """
        # YOUR CODE HERE
        if(self.method =='normal'):
            self.fit_normal_equation(X, y)
        elif(self.method == 'gradient'):
            self.fit_gradient_descent(X, y)
        
    
    def predict(self, X):
        """
        Make predictions using the fitted model.
        
        Parameters:
        -----------
        X : numpy array
            Feature matrix (m, n)
        
        Returns:
        --------
        predictions : numpy array
            Predicted values (m,)
        """
        # YOUR CODE HERE
        X_b = self._add_bias_term(X)
        return X_b.dot(np.hstack(([self.bias], self.weights)))
        
    
    def score(self, X, y):
        """
        Calculate R² score (coefficient of determination).
        
        Parameters:
        -----------
        X : numpy array
            Feature matrix (m, n)
        y : numpy array
            True target values (m,)
        
        Returns:
        --------
        r2 : float
            R² score
        """
        # YOUR CODE HERE
        y_pred = self.predict(X)
        ss_total = np.sum((y - np.mean(y)) ** 2)
        ss_residual = np.sum((y - y_pred) ** 2)
        r2 = 1 - (ss_residual / ss_total)
        return r2
        
    
    @property
    def coef_(self):
        """Return the coefficients (weights)."""
        return self.weights
    
    @property
    def intercept_(self):
        """Return the intercept (bias)."""
        return self.bias
```

```python
def run_complete_analysis(filepath, target_column='median_house_value', test_size=0.25, random_state=42):
    """
    Complete end-to-end linear regression analysis pipeline.
    
    Parameters:
    -----------
    filepath : str
        Path to the CSV file
    target_column : str
        Name of the target variable column
    test_size : float
        Proportion of data to use for testing (default: 0.25)
    random_state : int
        Random seed for reproducibility (default: 42)
    
    Returns:
    --------
    dict : Dictionary containing all models, data splits, and results
    """
    # 1. Load Data
    print("\n[1/5] Loading data...")
    df = pd.read_csv(filepath)
    print(f"✓ Loaded {df.shape[0]} rows and {df.shape[1]} columns")
    
    # 2. Handle Missing Values
    print("\n[2/5] Handling missing values...")
    df_cleaned = df.copy()
    numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns
    missing_count = 0
    for col in numeric_cols:
        null_count = df_cleaned[col].isnull().sum()
        if null_count > 0:
            median_val = df_cleaned[col].median()
            df_cleaned[col].fillna(median_val, inplace=True)
            missing_count += null_count
    print(f"✓ Filled {missing_count} missing values with median")
    
    # 3. Prepare Data
    print("\n[3/5] Preparing data for modeling...")
    numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns.tolist()
    numeric_cols.remove(target_column)
    
    X = df_cleaned[numeric_cols]
    y = df_cleaned[target_column]
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    print(f"✓ Train set: {X_train.shape[0]} samples")
    print(f"✓ Test set: {X_test.shape[0]} samples")
    print(f"✓ Features: {len(numeric_cols)}")
    
    # 4. Train Models
    print("\n[4/5] Training models...")
    
    # Normal Equation Method
    print("\n--- Normal Equation Method ---")
    model_normal = LinearRegressionScratch(method='normal')
    model_normal.fit(X_train_scaled, y_train.values)
    
    r2_train_normal = model_normal.score(X_train_scaled, y_train.values)
    r2_test_normal = model_normal.score(X_test_scaled, y_test.values)
    
    print(f"Intercept: {model_normal.intercept_:.4f}")
    print(f"Coefficients: {model_normal.coef_[:3]}... (showing first 3)")
    print(f"R² (Train): {r2_train_normal:.4f}")
    print(f"R² (Test): {r2_test_normal:.4f}")
    
    # Gradient Descent Method
    print("\n--- Gradient Descent Method ---")
    model_gd = LinearRegressionScratch(method='gradient', learning_rate=0.01, n_iterations=1000)
    model_gd.fit(X_train_scaled, y_train.values)
    
    r2_train_gd = model_gd.score(X_train_scaled, y_train.values)
    r2_test_gd = model_gd.score(X_test_scaled, y_test.values)
    
    print(f"Intercept: {model_gd.intercept_:.4f}")
    print(f"Coefficients: {model_gd.coef_[:3]}... (showing first 3)")
    print(f"R² (Train): {r2_train_gd:.4f}")
    print(f"R² (Test): {r2_test_gd:.4f}")
    
    # 5. Visualize Results
    print("\n[5/5] Generating visualizations...")
    if len(model_gd.cost_history) > 0:
        plt.figure(figsize=(10, 5))
        plt.plot(model_gd.cost_history, color='steelblue', linewidth=2)
        plt.xlabel('Iteration', fontsize=12)
        plt.ylabel('Cost (MSE)', fontsize=12)
        plt.title('Gradient Descent Convergence', fontsize=14, fontweight='bold')
        plt.grid(True, alpha=0.3)
        plt.show()
    
    print("\n" + "="*60)
    print("✓ ANALYSIS COMPLETE!")
    print("="*60)
    
    # Return all components for further analysis if needed
    return {
        'data': {
            'X_train': X_train,
            'X_test': X_test,
            'y_train': y_train,
            'y_test': y_test,
            'X_train_scaled': X_train_scaled,
            'X_test_scaled': X_test_scaled,
            'feature_names': numeric_cols
        },
        'models': {
            'normal_equation': model_normal,
            'gradient_descent': model_gd
        },
        'scaler': scaler,
        'results': {
            'normal_r2_train': r2_train_normal,
            'normal_r2_test': r2_test_normal,
            'gd_r2_train': r2_train_gd,
            'gd_r2_test': r2_test_gd
        }
    }
```
